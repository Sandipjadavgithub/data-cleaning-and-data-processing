# Sactter plot

```python
x= data["BlowFlow"].head(10)
y= data["ChipRate"].head(10)

#plt.plot(x,y)
plt.scatter(x,y,color='r',marker="*",s=100,linewidth=2)
plt.xlabel("this is blowflow")
plt.ylabel("this is chiprate")
plt.show()
```

```python
x= data["BlowFlow"].head(10)
y= data["ChipRate"].head(10)

plt.pie(y,labels=x,autopct="%0.1f%%",colors=["green","yellow","gray","red","purple"])
plt.xlabel("this is blowflow data")
plt.ylabel("this is chiprate data")
plt.show()
```
